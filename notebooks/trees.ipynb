# Planted trees per year

```python
unique_years = set(planted_trees)
unique_years
```

```python
for y in unique_years:
    print(y, planted_trees.count(y))
```

```python
trees_per_yr = []
for y in unique_years:
    trees_per_yr.append((y, planted_trees.count(y)))
trees_per_yr
```

```python
trees_per_yr_counts = [planted_trees.count(y) for y in unique_years]
trees_per_yr_counts
```

## Average planted trees per year

```python
avg_trees = sum(trees_per_yr_counts)/len(trees_per_yr_counts)
avg_trees
```
